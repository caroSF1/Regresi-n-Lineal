```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files 
from sklearn.linear_model import LinearRegression
```

```python
EUA = pd.read_csv("California_EUA (Limpio).csv", encoding = "unicode_escape", index_col = 'Unnamed: 0')
Mexico = pd.read_csv("DF_Mexico (Limpio).csv", index_col = 'Unnamed: 0')
Espana = pd.read_csv("Girona_España (Limpio).csv", encoding = "unicode_escape", index_col = 'Unnamed: 0')
```

```python
def correlacion(df, dependiente, independiente): 
  df1 = pd.DataFrame()
  df1 = df.copy()
  Vars_Indep= df[independiente]
  Var_Dep= df[dependiente]
  model= LinearRegression()
  model.fit(X=Vars_Indep, y=Var_Dep)
  y_pred= model.predict(X=df[independiente])
  df1.insert(0, 'Predicciones', y_pred)
  colores = ['yellow', 'blue', 'red', 'green', 'black', 'orange']
  tam = len(independiente)
  tam
  for i in range(tam):
    plt.scatter(x=independiente[i], y=dependiente, color=colores[i], data=df1)
    #plt.scatter(x=independiente[i], y='Predicciones', color="red", data=df1)
  plt.show()
  modeloMat = model.__dict__
  modeloMat = pd.DataFrame(modeloMat)
  print("El modelo matemático es: y = ")
  for i in range(tam):
    print("+ %fx%i" % (modeloMat['coef_'].values[i], i+1))
  print("+ %f" % ( modeloMat['intercept_'].values[0]))
  coef_Deter=model.score(X=Vars_Indep, y=Var_Dep)
  print("El coeficiente de determinación es: ", coef_Deter)
  coef_Correl=np.sqrt(coef_Deter)
  print("El coeficiente de correlación es: ", coef_Correl)
  return (coef_Deter, coef_Correl)
```

```python
def correlaciones_df(df, nombre):
  df2 = pd.DataFrame()
  lista_coef_Deter = []
  lista_coef_Correl = []
  lista = ['Modelo 1', 'Modelo 2', 'Modelo 3']
  coef_Deter, coef_Correl = correlacion(df, 'host_acceptance_rate', ['host_response_rate', 'price', 'number_of_reviews', 'review_scores_rating'])
  lista_coef_Deter.append(coef_Deter)
  lista_coef_Correl.append(coef_Correl)
  coef_Deter, coef_Correl = correlacion(df, 'host_acceptance_rate', ['availability_365', 'host_response_rate', 'number_of_reviews', 'accommodates'])
  lista_coef_Deter.append(coef_Deter)
  lista_coef_Correl.append(coef_Correl)
  coef_Deter, coef_Correl = correlacion(df, 'host_acceptance_rate', ['number_of_reviews', 'review_scores_cleanliness', 'review_scores_communication'])
  lista_coef_Deter.append(coef_Deter)
  lista_coef_Correl.append(coef_Correl)
  df2[nombre] = lista
  df2['Coeficiente determinación'] = lista_coef_Deter
  df2['Coeficiente correlación'] = lista_coef_Correl
  return df2
```

##**Entire Room**

```python
EUA_entireroom_corr = correlaciones_df(EUA[EUA['room_type']=='Entire home/apt'], 'EUA')
```

```python
Mexico_entireroom_corr = correlaciones_df(Mexico[Mexico['room_type']=='Entire home/apt'], 'México')
```

```python
Espana_entireroom_corr = correlaciones_df(Espana[Espana['room_type']=='Entire home/apt'], 'España')
```

```python
paises = pd.concat([EUA_entireroom_corr, Mexico_entireroom_corr, Espana_entireroom_corr], axis = 1)
paises
```

Primeramente, para el tipo de habitación: “Entire Room” se pueden observar los siguientes puntos importantes:

- En cuestión del coeficiente de determinación, podemos decir que los valores no son mayores a 50% (son valores extremadamente bajos), lo que nos indica que el cálculo no logra modelar los datos con precisión en absoluto. Asimismo, los coeficientes de correlación no fueron lo que se esperaba, dado que los valores eran inferiores a un 50%, esto nos dice que no se logra cuantificar la intensidad de la relación lineal entre las variables en el análisis de correlación presentado.
- **El modelo 2** perteneciente a **México**, correspondiente a “Price” vs availability_365, host_response_rate, number_of_reviews, accommodates“, nos muestra un porcentaje de **36.76%** con respecto a la correlación, mismo porcentaje que muestra el valor mayor en comparación con los demás modelos.
- Gracias a los porcentajes bajos en relación a la correlación, se puede intuir que el precio puede que tenga o no una influencia en la decisión del usuario al rentar el inmueble.
